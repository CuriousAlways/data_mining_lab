```python
import random
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
set1 = [(random.uniform(50,100),random.uniform(70,100)) for i in range(50)]
set2 = [(random.uniform(70,150),random.uniform(105,150)) for i in range(50)]
set3 = [(random.uniform(120,200),random.uniform(170,250)) for i in range(50)]
data = set1+set2+set3
plt.scatter([d[0] for d in data],[d[1] for d in data])
```

```python
no_examples = len(data)
no_examples
```

```python
x_mean = sum([d[0] for d in data])/no_examples
x_mean
```

```python
y_mean = sum([d[1] for d in data])/no_examples
y_mean
```

```python
#covariance(x,y)
cov = sum([(data[i][0]-x_mean)*(data[i][1]-y_mean) for i in range(no_examples)])/no_examples
cov
```

```python
#variance(X)
var = sum([(d[0]-x_mean)**2 for d in data])/no_examples
var
```

```python
b1 = cov/var
b0 = y_mean - (b1*x_mean)
print(b0,b1)
```

```python
#predict for
point = 240
prediction = b0 + b1*point
```

```python
#ypred = b0 + b1*x
fig= plt.figure(figsize=(12,7))
axes= fig.add_axes([0.4,0.4,.8,.8])
axes.scatter([d[0] for d in data],[d[1] for d in data],label='data')
axes.plot([d[0] for d in data]+[point],[b0+b1*d[0] for d in data]+[prediction],color='orange',label='prediction_line')
axes.scatter(point,prediction,color='black',s=100,label='point')
plt.legend()
```
